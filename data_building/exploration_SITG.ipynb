## First touch

```python
import pandas as pd

pd.set_option('display.max_columns', None)

print('Hello world')
```

```python
df = pd.read_csv('./data/REG_ENTREPRISE_ETABLISSEMENT.csv', sep=';')

df.head()
```

```python
df_entrep = pd.read_csv('./data/target_swiss_clean.csv')

list_geneva = df_entrep[df_entrep['City'] == 'Genève']['Name'].to_numpy()

print(list_geneva)

print(len(list_geneva))
```

```python
list_geneva_full = df['RAISON_SOC_PARENT'].dropna().to_numpy()
```

```python
import numpy as np

print(len(np.intersect1d(list_geneva, list_geneva_full)))

intersection = np.intersect1d(list_geneva, list_geneva_full)
```

```python
df_intersection = df[df['RAISON_SOC_PARENT'].isin(intersection)]

# df_intersection.to_csv('./data/intersection_geneve.csv')

df_intersection.head()
```

```python
df[df['NOM'].apply(lambda x: 'ALANA' in x)]['NOM']
```

```python
branche_list = df_intersection['BRANCHE'].unique()

print(branche_list)

df[df['BRANCHE'].isin(branche_list)]['BRANCHE'].value_counts()
```

```python
df_societes_invest = df[df['BRANCHE'] == "Sociétés d'investissement"]

print(f'The number of societies is {len(df_societes_invest)}')
print(f'The number of website is  {df_societes_invest['SITE_INTERNET'].notna().sum()}')
print(f'The number of email is {df_societes_invest['EMAIL'].notna().sum()}')
```

```python
df_societes_invest['IMMAT_DT'].apply(lambda x: x.split('-')[0]).value_counts().sort_index(ascending=False).plot()
```

```python
df[df['BRANCHE'].isin(branche_list)][['CODE_NOGA', 'BRANCHE']].drop_duplicates()
```

```python
list_code_noga = [
    663002,
    649901,
    643000   
]

df_final = df[df['CODE_NOGA'].isin(list_code_noga)]

df_final.head()

df_final.to_csv('./data/geneva_full_prospects.csv', index=False)
```
