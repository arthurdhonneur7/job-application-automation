## Getting the information for Geneva Based firms

```python
df.head()
```

```python
df_geneva = df[df['City'] == 'Genève']
```

```python
names_to_drop = [
    'Black Oak Family Office SA',
    'ATHENEE INVESTMENT SA'
]

df_geneva = df_geneva[~df_geneva['Name'].isin(names_to_drop)]
```

```python
dict_results = {k:"" for k in df_geneva['Name'].values}
```

### Doing the first 100


```python
from serpapi import GoogleSearch

api_key = ''


with open('geneva_names.txt', 'a') as fp:

    for name in df_geneva['Name'].values[:99]:
        
        try:
            search = GoogleSearch({
            "q": f"{name} Geneva official website",
            "api_key": api_key
            })

            res = search.get_dict()
            dict_results[name] = res
            
            fp.write(name + '\n')

        except:
            print(f"An error has occured with the company {name}")

        
    
```

## Fetching the data for zug firms

```python
df_zug = df[df['City'] == 'Zug']

dict_results = {k:"" for k in df_zug['Name'].values}
```

```python
df_zug['Name'].values
```

```python
from serpapi import GoogleSearch

api_key = ''
count = 0

with open('./data/zug_names.txt', 'a') as fp:

    for name in df_zug['Name'].values:
        
        try:
            count+= 1

            if count%10 == 0:
                print(f"We have fetched {count} companies")

            search = GoogleSearch({
            "q": f"{name} Zug official website",
            "api_key": api_key
            })

            res = search.get_dict()
            dict_results[name] = res
            
            fp.write(name + '\n')

        except:
            print(f"An error has occured with the company {name}")

print(f'We have fetched {count} searches')
    
```
